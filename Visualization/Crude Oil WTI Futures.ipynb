```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.transforms as mtransforms
import seaborn as sns
plt.rc('font',family='Times New Roman')

import datetime as dt
from datetime import timedelta
```

```python
name = ['Crude Oil WTI Futures.csv', 'Gold Futures.csv', 'SP500.csv', 'EURONEXT.csv', 'HANGSENG.csv', 'KOSPI.csv', 'NIKKEI.csv', 'SENSEX.csv', 'SSE.csv']
prices_1 = pd.read_csv('CrudeOil.csv', na_values='-',infer_datetime_format=True,squeeze=True).fillna(method ='pad')
```

```python
class draw_graph:
	def __init__(self, list_of_stock_prices, time, name):
		self.list_of_stock_prices = list_of_stock_prices
		self.time = time
		self.name = name
	def draw(self):
		plt.plot(self.time, self.list_of_stock_prices, color='mediumblue', linewidth=0.9)
		plt.tight_layout()
		plt.gcf().subplots_adjust(top=0.9, bottom=0.13, left=0.15)
		plt.ylabel('\nЦіна')
		plt.xlabel('Дата')
		index_name=''
		for i in range(len(self.name)-4):
			index_name = index_name + self.name[i]
		plt.title('Динаміка '+index_name)
		ax = plt.axes()
		ax.set_xlim(left=self.time[0], right=self.time[-1])
		ax.spines["right"].set_visible(False)
		ax.spines["top"].set_visible(False)
		plt.show()
# 		plt.savefig(index_name)
		plt.close()
```

```python
def save_few_graphs(price, time, name):
	c = draw_graph(price, time, name)
	c.draw()

class Time_list_month:
	def __init__(self, list_of_stock_prices):
		self.list_of_stock_prices = list_of_stock_prices

	def time_return(self):
		some_day = dt.datetime(year=1999, month=12, day=26)
		time = []
		for i in range(len(self.list_of_stock_prices)):
			some_day += timedelta(days=7)
			time.append(some_day)
		return time 
```

```python
time = Time_list_month(prices_1['Close'])
time = time.time_return()

save_few_graphs(prices_1['Close'], time, name[0])
```
